Task 9: Iterators and Generators
Create an iterator for the first 5 even numbers.Then, create a generator for the same.

```python
def g_even():
  for x in range(1,6):
    yield x*2

g_num = g_even()
for even_num in g_num:
  print(even_num)
```

Task 10: Map, Reduce, and Filter
Use map to square all numbers in a list.
Use reduce to find the product of all numbers in a list.Use filter to get only even numbers from a list.

```python
from functools import reduce
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x**2, numbers))
print("the square of numbers:",squared_numbers)

product = reduce(lambda x, y: x*y, numbers)
print(product)
even_num = list(filter(lambda x: x%2==0, numbers))
print(even_num)
```

 Task 11: Object-Oriented Programming - Creating a Class
Create a class called 'Rectangle' with attributes 'length' and 'width'.Include methods to calculate area and perimeter.


```python
class Rectangle:
  def __init__(self, length, width):
    self.length = length
    self.width = width

  def area(self):
    return self.length * self.width

  def perimeter(self):
    return 2 * self.length + 2 * self.width

x = Rectangle(3,4)
print(x.area())
print(x.perimeter())
```

Task 12: Pandas Data Analysis
Using the following DataFrame, perform these tasks:1. Find the average salary by department
Get the names of employees with salary > 60000
Add a new column 'Bonus' which is 10% of salary

```python
import pandas as pd
df_employees = pd.DataFrame({
    'Name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [55000, 65000, 70000, 60000, 58000]})

avg_salary = df_employees.groupby('Department')['Salary'].mean()
print(avg_salary)
high_salary = df_employees[df_employees['Salary'] > 60000]
print(high_salary)
df_employees['Bonus'] = 0.1 * df_employees['Salary']
print(df_employees)
```
